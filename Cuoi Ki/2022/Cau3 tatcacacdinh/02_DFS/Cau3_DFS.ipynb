Dán code:

````python
V = ["S", "A", "B", "C", "D", "E", "F", "G", "H", "K", "M", "N", "I", "J", "L", "Z"]

H = {
    "S": 10,
    "A": 9,
    "B": 8,
    "C": 7,
    "D": 6,
    "E": 5,
    "F": 4,
    "G": 10,
    "H": 10,
    "K": 3,
    "M": 0,
    "N": 10,
    "I": 6,
    "J": 0,
    "L": 9,
    "Z": 8,
}

E = [
    ("S", "A", 5),
    ("S", "B", 6),
    ("S", "C", 5),
    ("A", "D", 6),
    ("A", "E", 7),
    ("D", "M", 5),
    ("D", "N", 8),
    ("E", "I", 8),
    ("B", "F", 3),
    ("B", "G", 4),
    ("F", "J", 4),
    ("F", "L", 4),
    ("C", "H", 6),
    ("C", "K", 4),
    ("K", "Z", 2),
]

class Graph:
    def __init__(self):
        self.adjacency = {}
        self.heuristic = {}

    def add_node(self, node, heuristic_value):
        if node not in self.adjacency:
            self.adjacency[node] = []
        self.heuristic[node] = heuristic_value

    def add_edge(self, u, v, cost):
        self.adjacency[u].append((v, cost))
        self.adjacency[v].append((u, cost))

    def sort_neighbors(self, vertex_order):
        for node in self.adjacency:
            self.adjacency[node].sort(key=lambda item: vertex_order[item[0]])


def build_graph(vertices, edges, heuristic_values):
    graph = Graph()

    for vertex in vertices:
        graph.add_node(vertex, heuristic_values[vertex])

    for u, v, cost in edges:
        graph.add_edge(u, v, cost)

    vertex_order = {vertex: index for index, vertex in enumerate(vertices)}
    graph.sort_neighbors(vertex_order)
    return graph


def format_adjacency(graph):
    lines = []

    for vertex in V:
        neighbors = ", ".join(
            f"{neighbor}(cost={cost})"
            for neighbor, cost in graph.adjacency[vertex]
        )
        lines.append(f"{vertex}: {neighbors}")

    return "\n".join(lines)


def main():
    graph = build_graph(V, E, H)

    print("Tap dinh V:")
    print(V)
    print()
    print("Heuristic H:")
    print(H)
    print()
    print("Tap canh E:")
    print(E)
    print()
    print("Danh sach ke:")
    print(format_adjacency(graph))


if __name__ == "__main__":
    main()
````

### Giải thích câu a

Ở câu a, em biểu diễn cây bằng ba thành phần chính:

- `V` là danh sách toàn bộ các đỉnh trong cây.
- `H` là bảng trọng số heuristic trên đỉnh. Đỉnh đích là đỉnh có `h = 0`, trong bài này gồm `M` và `J`.
- `E` là danh sách cạnh, mỗi cạnh có dạng `(đỉnh_1, đỉnh_2, chi_phí)`.

Vì cây trong bài được xem là vô hướng nên khi thêm cạnh, chương trình thêm cả hai chiều vào danh sách kề. Ví dụ cạnh `("S", "A", 5)` cho biết có thể đi từ `S` sang `A` và từ `A` về `S`, với chi phí cạnh bằng `5`.

Class `Graph` lưu hai thông tin quan trọng:

- `adjacency` lưu danh sách kề của từng đỉnh, kèm chi phí cạnh.
- `heuristic` lưu giá trị `h(n)` của từng đỉnh.

Sau khi khởi tạo xong cây, chương trình in ra tập đỉnh, bảng heuristic, tập cạnh và danh sách kề để kiểm tra dữ liệu đã đúng với đề bài.


b)	Hãy viết chương trình sử dụng thuật toán DFS để tìm đường đi từ đỉnh `S` đến tất cả các đỉnh có `h = 0`. Trong chương trình, hãy in ra thứ tự đỉnh khám phá trong quá trình tìm kiếm và bảng trạng thái ngăn xếp.

Dán code:

````python


V = ["S", "A", "B", "C", "D", "E", "F", "G", "H", "K", "M", "N", "I", "J", "L", "Z"]

H = {
    "S": 10, "A": 9, "B": 8, "C": 7, "D": 6, "E": 5, "F": 4, "G": 10,
    "H": 10, "K": 3, "M": 0, "N": 10, "I": 6, "J": 0, "L": 9, "Z": 8,
}

E = [
    ("S", "A", 5), ("S", "B", 6), ("S", "C", 5), ("A", "D", 6),
    ("A", "E", 7), ("D", "M", 5), ("D", "N", 8), ("E", "I", 8),
    ("B", "F", 3), ("B", "G", 4), ("F", "J", 4), ("F", "L", 4),
    ("C", "H", 6), ("C", "K", 4), ("K", "Z", 2),
]


class Graph:
    def __init__(self):
        self.adjacency = {}
        self.heuristic = {}

    def add_node(self, node, heuristic_value):
        if node not in self.adjacency:
            self.adjacency[node] = []
        self.heuristic[node] = heuristic_value

    def add_edge(self, u, v, cost):
        self.adjacency[u].append((v, cost))
        self.adjacency[v].append((u, cost))

    def sort_neighbors(self, vertex_order):
        for node in self.adjacency:
            self.adjacency[node].sort(key=lambda item: vertex_order[item[0]])


def build_graph(vertices, edges, heuristic_values):
    graph = Graph()
    for vertex in vertices:
        graph.add_node(vertex, heuristic_values[vertex])
    for u, v, cost in edges:
        graph.add_edge(u, v, cost)
    vertex_order = {vertex: index for index, vertex in enumerate(vertices)}
    graph.sort_neighbors(vertex_order)
    return graph


def reconstruct_path(parent, start, goal):
    path = []
    current = goal
    while current is not None:
        path.append(current)
        current = parent[current]
    path.reverse()
    return path if path[0] == start else None


def path_cost(path, graph):
    total = 0
    for u, v in zip(path, path[1:]):
        for neighbor, cost in graph.adjacency[u]:
            if neighbor == v:
                total += cost
                break
    return total


def dfs_to_zero_heuristic_goals(graph, start):
    stack = [start]
    visited = set()
    discovered = {start}
    parent = {start: None}
    exploration_order = []
    steps = []
    goals = []

    while stack:
        stack_before = list(stack)
        current = stack.pop()
        if current in visited:
            continue
        visited.add(current)
        exploration_order.append(current)

        if graph.heuristic[current] == 0:
            path = reconstruct_path(parent, start, current)
            goals.append({
                "goal": current,
                "path": path,
                "cost": path_cost(path, graph),
            })
            steps.append({
                "stack_before": stack_before,
                "current": current,
                "h": graph.heuristic[current],
                "added": [],
                "stack_after": list(stack),
                "note": "Ghi nhan dinh dich h = 0",
            })
            continue

        added = []
        for neighbor, _ in reversed(graph.adjacency[current]):
            if neighbor not in discovered:
                discovered.add(neighbor)
                parent[neighbor] = current
                stack.append(neighbor)
                added.append(neighbor)

        steps.append({
            "stack_before": stack_before,
            "current": current,
            "h": graph.heuristic[current],
            "added": added,
            "stack_after": list(stack),
            "note": "",
        })

    return goals, exploration_order, steps


def format_list(values):
    return "[" + ", ".join(values) + "]"


def make_table(headers, rows):
    widths = [
        max(len(str(row[index])) for row in [headers] + rows)
        for index in range(len(headers))
    ]
    separator = "+" + "+".join("-" * (width + 2) for width in widths) + "+"

    def format_row(row):
        cells = [
            str(value).ljust(widths[index])
            for index, value in enumerate(row)
        ]
        return "| " + " | ".join(cells) + " |"

    lines = [separator, format_row(headers), separator]
    lines.extend(format_row(row) for row in rows)
    lines.append(separator)
    return "\n".join(lines)


def format_steps(steps):
    headers = ["Buoc", "Stack truoc", "Lay ra", "Them vao", "Stack sau"]
    rows = []
    for index, step in enumerate(steps, start=1):
        added = ", ".join(step["added"]) if step["added"] else "Khong co"
        rows.append([
            index,
            format_list(step["stack_before"]),
            f"{step['current']}(h={step['h']})",
            added,
            format_list(step["stack_after"]),
        ])
    return "\n".join([
        "Bang trang thai:",
        make_table(headers, rows),
        "Ghi nhan moi dinh co h = 0 va tiep tuc cho den khi stack rong.",
    ])


def solve():
    graph = build_graph(V, E, H)
    goals, exploration_order, steps = dfs_to_zero_heuristic_goals(graph, "S")
    lines = [
        "THUAT TOAN DFS",
        "Thu tu dinh kham pha:",
        " -> ".join(exploration_order),
        "",
        format_steps(steps),
        "",
    ]
    if not goals:
        lines.append("Khong tim thay duong di")
    else:
        lines.append("Cac dinh dich tim duoc:")
        lines.append(", ".join(item["goal"] for item in goals))
        lines.append("Cac duong di tu S den cac dinh co h = 0:")
        for item in goals:
            lines.append(f'{item["goal"]}: { " -> ".join(item["path"]) } | Tong chi phi = {item["cost"]}')
    return "\n".join(lines)


print(solve())
````

### Giải thích tác dụng của từng hàm

- `Graph.__init__()`: khởi tạo cây rỗng, gồm `adjacency` để lưu danh sách kề và `heuristic` để lưu giá trị `h(n)` của từng đỉnh.
- `Graph.add_node(node, heuristic_value)`: thêm một đỉnh vào cây và lưu giá trị heuristic của đỉnh đó. Nếu đỉnh chưa có trong danh sách kề thì tạo danh sách kề rỗng cho đỉnh.
- `Graph.add_edge(u, v, cost)`: thêm cạnh có chi phí giữa hai đỉnh `u` và `v`. Vì cây được biểu diễn dạng vô hướng nên cạnh được lưu ở cả hai chiều.
- `Graph.sort_neighbors(vertex_order)`: sắp xếp các đỉnh kề theo thứ tự xuất hiện trong tập đỉnh `V`, giúp thứ tự duyệt ổn định và dễ kiểm tra.
- `build_graph(vertices, edges, heuristic_values)`: tạo đối tượng `Graph`, thêm toàn bộ đỉnh, thêm toàn bộ cạnh, sắp xếp danh sách kề rồi trả về cây hoàn chỉnh.
- `reconstruct_path(parent, start, goal)`: truy vết đường đi từ đỉnh đích về đỉnh bắt đầu dựa trên dictionary `parent`, sau đó đảo ngược để lấy đường đi đúng chiều từ `start` đến `goal`.
- `path_cost(path, graph)`: tính tổng chi phí của một đường đi bằng cách cộng chi phí các cạnh liên tiếp trong `path`.
- `format_list(values)`: chuyển một danh sách giá trị thành chuỗi để in ra màn hình theo định dạng dễ đọc.
- `make_table(headers, rows)`: tạo bảng văn bản từ tiêu đề cột và các dòng dữ liệu.
- `format_row(row)`: hàm phụ dùng khi tạo bảng, giúp căn chỉnh dữ liệu của từng dòng.
- `format_steps(steps)`: chuyển danh sách các bước chạy thuật toán thành bảng trạng thái để người đọc theo dõi quá trình duyệt.
- `dfs_to_zero_heuristic_goals(graph, start)`: thực hiện DFS từ `start` để tìm tất cả các đỉnh có `h = 0`. Hàm dùng stack, đi sâu theo từng nhánh, lưu thứ tự khám phá, lưu quan hệ cha và tiếp tục duyệt cho đến khi stack rỗng.
- `solve()`: xây dựng cây, chạy DFS, gom bảng bước chạy, thứ tự khám phá, các đường đi tìm được và tổng chi phí thành chuỗi kết quả hoàn chỉnh.

Ở câu b, em sử dụng thuật toán DFS để tìm đường đi từ `S` đến tất cả các đỉnh có heuristic bằng `0`.

DFS dùng ngăn xếp `stack`, nên thuật toán đi sâu theo một nhánh trước khi quay lui. Trong chương trình, các đỉnh kề được duyệt bằng `reversed(...)` để khi đưa vào stack, đỉnh được ưu tiên theo thứ tự khai báo trong `V` sẽ nằm trên cùng và được lấy ra trước.

### Công thức / nguyên tắc sử dụng

DFS không dùng công thức chi phí hay heuristic để chọn đỉnh tiếp theo. Thuật toán dựa trên nguyên tắc vào sau ra trước của ngăn xếp:

```text
LIFO = Last In, First Out
```

Điều kiện ghi nhận đích của bài này là:

```text
h(current) = 0
```

Khi DFS lấy ra một đỉnh có trọng số heuristic bằng `0`, thuật toán lưu lại đỉnh đó vào `goals`, truy vết đường đi bằng `parent`, rồi tiếp tục duyệt cho đến khi `stack` rỗng.

Các biến chính trong chương trình:

- `stack` là ngăn xếp chứa các đỉnh chờ duyệt.
- `visited` là tập các đỉnh đã được khám phá.
- `discovered` là tập các đỉnh đã từng được đưa vào stack.
- `parent` lưu cha của mỗi đỉnh để truy vết lại đường đi.
- `goals` lưu toàn bộ các đỉnh có `h = 0` tìm được.
- `exploration_order` lưu thứ tự các đỉnh được lấy ra khỏi stack.
- `steps` lưu trạng thái stack trước và sau mỗi bước để lập bảng chạy chi tiết.

DFS tìm được các đỉnh đích theo thứ tự duyệt sâu. Thứ tự này phụ thuộc vào cách sắp xếp danh sách kề và cách đưa đỉnh vào stack.
